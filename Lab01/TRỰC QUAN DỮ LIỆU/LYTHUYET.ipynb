Làm thế nào để tạo một biểu đồ đơn giản như histogram hoặc bar chart bằng Matplotlib? Bạn có thể chia sẻ đoạn code mẫu không?

### 1. Nạp thư viện

```python

import matplotlib.pyplot as plt
import numpy as np
```

### 2. Tải dữ liệu mẫu

```python
data = np.random.randn(1000)  # Dữ liệu ngẫu nhiên cho histogram
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 67]     # Dữ liệu cho bar chart
```

### 3. Tạo histogram

```python
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Sample')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

### 4. Tạo bar chart

```python
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='lightgreen', edgecolor='black')
plt.title('Bar Chart Sample')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```
