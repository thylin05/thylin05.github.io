Viết đoạn code mẫu bằng Python (sử dụng Scikit-learn) để triển khai K-Means Clustering không? Hãy mô tả các bước thực hiện

### 1. Nạp thư viện

```python
from sklearn.cluster import KMeans
import numpy as np
```

### 2. Tải dữ liệu mẫu

```python
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])
```

### 3. Khởi tạo và huấn luyện mô hình K-Means với K=2

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
```

### 4. Lấy nhãn cụm và tâm cụm

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

### 5. In kết quả

```python
print("Nhãn cụm:", labels)
print("Tâm cụm:", centroids)
```

# CÂU 6

Sử dụng phương pháp nào trong Python để chọn số cụm K tối ưu? Hãy chia sẻ một đoạn code mẫu

### 1. Nạp thư viện

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

### 2. Dữ liệu mẫu

```python
X = np.array([[1, 2], [1, 4], [1, 0], [10, 2], [10, 4], [10, 0]])
```

### 3. Tính WCSS (Within-Cluster Sum of Squares) cho các giá trị K

```python
wcss = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

### 4. Vẽ biểu đồ Elbow

```python
plt.plot(range(1, 6), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Số cụm (K)')
plt.ylabel('WCSS')
plt.show()
```
